```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
cars = pd.read_csv('/content/drive/MyDrive/Projetos/JACTO/CarsData.csv')
total_cars = len(cars['year'])
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
cars.groupby('transmission').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

#Análise 1
Ao classificarmos carros por ano de fábricação é possível perceber que 99.01% (96.749) foram fábricados entre 2010 e 2020.

Indo um pouco mais afundo, vemos que existem apenas dois carros com data de fábricação de 1970, e o mais antigo, fora eles, foi fábricado em 1996.

Também é possível notar que existe apenas um carro com data de fábricação após 2020, o Ford Fiesta (2024).


```python
# Plotar carros por ano de fábricação.
cars['year'].plot(kind='hist', bins=20, title='Carros por ano de fábricação', color=sns.palettes.mpl_palette('Dark2'))

for bar in plt.gca().patches:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(int(bar.get_height())),
             ha='center', va='bottom', fontsize=8)

plt.show()

# Calcular porcentagem dos carros fábricados entre 2010 e 2020
cars_filter_year = len(cars[(cars['year'] >= 2010) & (cars['year'] <= 2020)])

print(f'\nPorcentagem dos carros fábricados entre 2010 e 2020: {((cars_filter_year / total_cars) * 100):.2f}% ({cars_filter_year})')

sorted_cars = cars.sort_values(by='year')

old_cars = sorted_cars.head(5)
print('\nLista dos 5 carros mais antigos\n', old_cars[['model', 'price', 'transmission', 'fuelType', 'Manufacturer', 'year']].to_string(index=False))

new_cars = cars[cars['year'] > 2020]
print('\nO carro mais novo dessa lista é o:\n', new_cars[['model', 'price', 'transmission', 'fuelType', 'Manufacturer', 'year']].to_string(index=False))
```

#Análise 2

Classificando os carros por tipo de transmissão também é possível notar que 55.25% utilizam Gasolina Comum como combustível e cerca de 41.37%

```python
fuel_type_counts = cars.groupby('fuelType').size()
fuel_type_counts.plot(kind='barh', title='Quantidade de carros por tipo de combustível', color=sns.palettes.mpl_palette('Dark2'))


for i, v in enumerate(fuel_type_counts):
    percentage = (v / total_cars) * 100
    plt.text(v, i, f'{percentage:.2f}%', va='center')

plt.show()
```
