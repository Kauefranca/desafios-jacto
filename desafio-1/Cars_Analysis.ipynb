```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
cars = pd.read_csv('/content/drive/MyDrive/Projetos/JACTO/CarsData.csv')
total_cars = len(cars['year'])
```

```python
cars
```

```python
# Carros por ano de fÃ¡bricaÃ§Ã£o.
cars['year'].plot(kind='hist', bins=20, title='Carros por ano de fÃ¡bricaÃ§Ã£o', color=sns.palettes.mpl_palette('Dark2'))

for bar in plt.gca().patches:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(int(bar.get_height())),
             ha='center', va='bottom', fontsize=8)

plt.show()
```

```python
# Carros por Tipo de CombustÃ­vel.
fuel_type_counts = cars.groupby('fuelType').size()
fuel_type_counts.plot(kind='barh', title='Carros por Tipo de CombustÃ­vel', color=sns.palettes.mpl_palette('Dark2'))

for i, v in enumerate(fuel_type_counts):
    percentage = (v / total_cars) * 100
    plt.text(v, i, f'{percentage:.2f}%', va='center')

plt.show()
```

```python
# DistribuiÃ§Ã£o de preÃ§os
plt.figure(figsize=(10, 6))
sns.histplot(cars['price'], bins=30, kde=True)
plt.title('DistribuiÃ§Ã£o de PreÃ§os')
plt.xlabel('PreÃ§o')
plt.ylabel('Contagem')

plt.show()
```

```python
# RelaÃ§Ã£o entre preÃ§o e ano do modelo
plt.figure(figsize=(10, 6))
sns.scatterplot(data=cars, x='year', y='price')
plt.title('RelaÃ§Ã£o entre PreÃ§o e Ano do Modelo')
plt.xlabel('Ano do Modelo')
plt.ylabel('PreÃ§o')

plt.show()
```

```python
# MPG por Tipo de CombustÃ­vel.
media_mpg = cars.groupby('fuelType')['mpg'].mean().reset_index()

plt.figure(figsize=(10, 6))
ax = sns.barplot(data=media_mpg, x='fuelType', y='mpg', hue='fuelType', palette=sns.color_palette('dark', 5))
plt.title('MPG por Tipo de CombustÃ­vel Utilizado')
plt.xlabel('Tipo de CombustÃ­vel')
plt.ylabel('MPG')

for p in ax.patches:
    ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

```python
# Filtro para selecionar apenas os carros a diesel e gasolina.
cars_filtered = cars[cars['fuelType'].isin(['Diesel', 'Petrol']) & (cars['engineSize'] > 0)]

media_por_cilindrada = cars_filtered.groupby('engineSize')['mpg'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.regplot(data=media_por_cilindrada, x='engineSize', y='mpg', line_kws={"color": "red"})
plt.title('MÃ©dia de Consumo por Cilindrada (Gasolina e Diesel)')
plt.xlabel('Cilindradas')
plt.ylabel('MÃ©dia de Consumo de CombustÃ­vel (MPG)')
plt.grid(True)

plt.show()
```

```python
# Filtrar carros para compara-los atravÃ©s dos anos.
comp = cars[((cars['engineSize'] >= 1.0) | (cars['engineSize'] <= 2.0)) & (cars['year'] >= 2010) & (cars['year'] != 2024)]
```

```python
years = sorted(comp['year'].unique())

price_per_year = {}

for year in years:
  price_per_year[year] = (comp[comp['year'] == year]['price'].mean())

deprec_by_year = {}

anos = list(price_per_year.keys())

for i in range(len(anos) - 1):
    curr_y = anos[i]
    next_y = anos[i + 1]
    deprec = price_per_year[curr_y] - price_per_year[next_y]
    deprec_by_year[next_y] = deprec

print(deprec_by_year)
```

```python
anos = list(deprec_by_year.keys())
depreciacoes = list(deprec_by_year.values())

plt.figure(figsize=(10, 6))
plt.bar(anos, depreciacoes, color='lightgreen')
plt.title('DepreciaÃ§Ã£o por Ano do Modelo')
plt.xlabel('Ano')
plt.ylabel('PreÃ§o ($)')

for i in range(len(anos)):
    plt.text(anos[i], depreciacoes[i] - 300, '{:.2f}$'.format(depreciacoes[i]), ha='center', va='bottom', fontsize=9)

plt.show()
```
